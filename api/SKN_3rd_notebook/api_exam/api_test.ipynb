```python
import requests
import pprint
import json

url = "http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getUltraSrtFcst?serviceKey=5T4olJF026%2B%2BfuPx%2FVB7BMtphjcDngrzt5lOgeK7p4EmhYBzP660Qv24QBmAyK3GSvatp8HOVAAgtgCcBNOdgA%3D%3D&pageNo=1&numOfRows=1000&dataType=json&base_date=20240731&base_time=0630&nx=55&ny=127"

response = requests.get(url)

contents = response.text

print(contents)
```

```python
pp = pprint.PrettyPrinter(indent=5)
print(pp.pprint(contents))
```

```python
#문자열을 json으로 변경
json_ob = json.loads(contents)
print(json_ob)
print(type(json_ob)) #json타입 확인
```

```python
# 필요한 내용만 꺼내기
body = json_ob['response']['body']['items']['item']
print(body)
```

```python
# pandas import
import pandas as pd

# Dataframe으로 만들기
dataframe = pd.json_normalize(body)

print(dataframe)
```
